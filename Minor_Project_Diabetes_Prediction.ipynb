```
# Importing Dependencies
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm

# Loading the dataset
diabetes_dataset = pd.read_csv('/content/drive/MyDrive/Diabetes_Dataset/diabetes.csv')

# Inspecting the dataset
print("First 5 rows of the dataset:")
print(diabetes_dataset.head())
print("\nDataset Shape:", diabetes_dataset.shape)
print("\nOutcome Counts:\n", diabetes_dataset['Outcome'].value_counts())

# Separating the features (X) and labels (Y)
X = diabetes_dataset.drop(columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']

# Data Standardization
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

print("\nData Shapes:")
print("X:", X.shape)
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)

# Initializing the models
models = {
    "Support Vector Machine (SVM)": svm.SVC(kernel='linear'),
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

# Training, Evaluating, and Comparing Models
results = {}

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, Y_train)

    # Predictions on training data
    train_predictions = model.predict(X_train)
    train_accuracy = accuracy_score(Y_train, train_predictions)

    # Predictions on test data
    test_predictions = model.predict(X_test)
    test_accuracy = accuracy_score(Y_test, test_predictions)

    # Store the results
    results[model_name] = {
        "Training Accuracy": train_accuracy,
        "Test Accuracy": test_accuracy
    }

# Displaying model performances
print("\nModel Performance Summary:\n")
for model_name, metrics in results.items():
    print(f"{model_name}:")
    print(f"  Training Accuracy: {metrics['Training Accuracy']:.4f}")
    print(f"  Test Accuracy: {metrics['Test Accuracy']:.4f}")
    print("-" * 30)

# Example Predictive System with SVM
input_data = (5, 166, 72, 19, 175, 25.8, 0.587, 51)

# Converting input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping and standardizing the input data
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
std_data = scaler.transform(input_data_reshaped)

# Making a prediction using SVM
svm_model = models["Support Vector Machine (SVM)"]
prediction = svm_model.predict(std_data)
print("\nPrediction for input data using SVM:")
if prediction[0] == 0:
    print("The person is not diabetic.")
else:
    print("The person is diabetic.")
```

```
# Importing Dependencies
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from imblearn.over_sampling import SMOTE  # For synthetic data generation

# Loading the dataset
diabetes_dataset = pd.read_csv('/content/drive/MyDrive/Diabetes_Dataset/diabetes.csv')

# Inspecting the dataset
print("First 5 rows of the dataset:")
print(diabetes_dataset.head())
print("\nDataset Shape:", diabetes_dataset.shape)
print("\nOutcome Counts:\n", diabetes_dataset['Outcome'].value_counts())

# Separating the features (X) and labels (Y)
X = diabetes_dataset.drop(columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']

# Data Standardization
scaler = StandardScaler()
X = scaler.fit_transform(X)


# Augmenting Data Using SMOTE to Handle Class Imbalance
print("\nBefore SMOTE:")
print("Class Distribution:\n", pd.Series(Y).value_counts())

smote = SMOTE(random_state=2)
X_resampled, Y_resampled = smote.fit_resample(X, Y)

print("\nAfter SMOTE:")
print("Class Distribution:\n", pd.Series(Y_resampled).value_counts())

# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_resampled, Y_resampled, test_size=0.2, stratify=Y_resampled, random_state=2)

print("\nData Shapes After Resampling:")
print("X_resampled:", X_resampled.shape)
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)

# Initializing the models
models = {
    "Support Vector Machine (SVM)": svm.SVC(kernel='linear'),
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

# Training, Evaluating, and Comparing Models
results = {}

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, Y_train)

    # Predictions on training data
    train_predictions = model.predict(X_train)
    train_accuracy = accuracy_score(Y_train, train_predictions)

    # Predictions on test data
    test_predictions = model.predict(X_test)
    test_accuracy = accuracy_score(Y_test, test_predictions)

    # Store the results
    results[model_name] = {
        "Training Accuracy": train_accuracy,
        "Test Accuracy": test_accuracy
    }

# Displaying model performances
print("\nModel Performance Summary:\n")
for model_name, metrics in results.items():
    print(f"{model_name}:")
    print(f"  Training Accuracy: {metrics['Training Accuracy']:.4f}")
    print(f"  Test Accuracy: {metrics['Test Accuracy']:.4f}")
    print("-" * 30)

# Example Predictive System with SVM
input_data = (5, 166, 72, 19, 175, 25.8, 0.587, 51)

# Converting input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping and standardizing the input data
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
std_data = scaler.transform(input_data_reshaped)

# Making a prediction using SVM
svm_model = models["Support Vector Machine (SVM)"]
prediction = svm_model.predict(std_data)
print("\nPrediction for input data using SVM:")
if prediction[0] == 0:
    print("The person is not diabetic.")
else:
    print("The person is diabetic.")
```
